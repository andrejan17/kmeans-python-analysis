```python
#Import libary yang dibutukan
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
#Membaca dataset dari file
df = pd.read_csv("D:/dataset_sebelum_preprocessing.csv")
df.info()
df.describe()
```

```python
#Preprocessing Data
#1. Pembersihan Data
df.isnull().sum()
```

```python
df1 = df.dropna()
df1
```

```python
df1.duplicated().sum()
```

```python
df1.loc[df1.duplicated(), :]
```

```python
df2 = df1.drop_duplicates(subset=['Account'])
df2
```

```python
#2. Seleksi Data
df3 = df2.drop(["Matches"], axis = 1)
df3.info()
```

```python
#3. Transformasi Data
df3["Favorite"].value_counts()
```

```python
transform_data = {"Favorite": {"Tank": 1, "Fighter": 2, "Assassin": 3, "Mage": 4,
                                  "Marksman": 5, "Support": 6}}
df3.replace(transform_data, inplace=True)
df3.info()
```

```python
print(df3["Favorite"].value_counts())
df3
```

```python
#4. Normalisasi Data
df4 = df3.iloc[:, 1:8]
df_array = np.array(df4)
df_array
```

```python
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_array)
np.set_printoptions(precision=2)
df_scaled
```

```python
pd.set_option('precision', 2)
df5 = pd.DataFrame(df_scaled, columns=['kda','teamfight','gold/min','damagetohero/min',
                                      'death/match','turretdamage/match','favorite'])
df5
```

```python
column_account = df3.drop(['KDA','Teamfight','Gold/Min','Damage to Hero/Min','Deaths/Match','Turret Damage/Match','Favorite'], axis=1)
column_array = np.array(column_account)
df6 = pd.DataFrame(column_array, columns=['account'])
df6
```

```python
dataset = df6.join(df5)
dataset
```

```python
dataset.to_csv('D:/dataset_setelah_preprocessing.csv')
```
